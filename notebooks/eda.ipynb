## Data Cleaning

```python
# CHECKING DUPLICATE DATA
df[df.duplicated(subset=df.columns.difference(['ID']))]
```

```python
# REMOVE DUPLICATE DATA
df = df.drop_duplicates(subset=df.columns.difference(['ID']))
```

```python
# DROP UNNECESSARY COLUMNS
columns_to_drop = ['FIELD4', 'Competitors', 'Company_Name', 'Size', 'Salary_Estimate', 'Headquarters', 'Revenue' ]
df.drop(columns=columns_to_drop, inplace=True, errors='ignore')
```

```python
# RENAME COLUMNS
df.rename(columns={ 'Job_Title': 'Jobs_Name', 'Location': 'Job_City', 'Founded': 'Company_Founded_Year',
                    'Hourly': 'Hourly_Salary', 'Employer_provided': 'Employer_Provided_Salary',
                    'Type_of_ownership': 'Type_Of_Ownership', 'company_txt': 'Company_Name',
                    'Avg_SalaryK': 'Salary_Avg', 'Lower_Salary': 'Salary_Lower', 'Upper_Salary': 'Salary_Upper',
                    'Job_Location': 'Job_State', 'Age': 'Company_Age', 'Python': 'Python',
                    'spark': 'Spark', 'aws': 'AWS', 'excel': 'Excel', 'sql_': 'SQL', 'sas': 'SAS',
                    'keras': 'Keras', 'pytorch': 'PyTorch', 'scikit': 'Scikit-learn', 'hadoop': 'Hadoop',
                    'tensor': 'TensorFlow', 'tableau': 'Tableau', 'bi': 'Power BI', 'flink': 'Flink',
                    'mongo': 'MongoDB', 'google_an': 'Google Analytics', 'job_title_sim': 'Job_Title_Sim',
                    'seniority_by_title': 'Seniority_Level', 'Degree': 'Employee_Degree' }, inplace=True)
```

```python
# CHANGE DATA TYPES 
df['Rating'] = df['Rating'].astype(float)
```

```python
# EXTRACT CITY NAME BEFORE COMMA
df['Job_City'] = df['Job_City'].astype(str).str.split(',').str[0].str.strip()
```

```python
# MAP STATE ABBREVIATIONS TO FULL NAMES
state_mapping = { 'AL': 'Alabama', 'AK': 'Alaska', 'AZ': 'Arizona', 'AR': 'Arkansas', 'CA': 'California',
                  'CO': 'Colorado', 'CT': 'Connecticut', 'DE': 'Delaware', 'FL': 'Florida', 'GA': 'Georgia',
                  'HI': 'Hawaii', 'ID': 'Idaho', 'IL': 'Illinois', 'IN': 'Indiana', 'IA': 'Iowa',
                  'KS': 'Kansas', 'KY': 'Kentucky', 'LA': 'Louisiana', 'ME': 'Maine', 'MD': 'Maryland',
                  'MA': 'Massachusetts', 'MI': 'Michigan', 'MN': 'Minnesota', 'MS': 'Mississippi',
                  'MO': 'Missouri', 'MT': 'Montana', 'NE': 'Nebraska', 'NV': 'Nevada', 'NH': 'New Hampshire',
                  'NJ': 'New Jersey', 'NM': 'New Mexico', 'NY': 'New York', 'NC': 'North Carolina',
                  'ND': 'North Dakota', 'OH': 'Ohio', 'OK': 'Oklahoma', 'OR': 'Oregon', 'PA': 'Pennsylvania',
                  'RI': 'Rhode Island', 'SC': 'South Carolina', 'SD': 'South Dakota', 'TN': 'Tennessee',
                  'TX': 'Texas', 'UT': 'Utah', 'VT': 'Vermont', 'VA': 'Virginia', 'WA': 'Washington',
                  'WV': 'West Virginia', 'WI': 'Wisconsin', 'WY': 'Wyoming', 'DC': 'District of Columbia' }

df['Job_State'] = df['Job_State'].astype(str).str.strip().str.upper()
df['Job_State'] = df['Job_State'].replace(state_mapping)
```

```python
# CONVERT SALARY FROM K TO FULL VALUE
df['Salary_Lower'] = df['Salary_Lower'] * 1000
df['Salary_Upper'] = df['Salary_Upper'] * 1000
df['Salary_Avg'] = df['Salary_Avg'] * 1000
```

```python
# CLEAN Job_Title_Sim 
job_title_mapping = { 'data scientist': 'Data Scientist', 'data scientist project manager': 'Data Scientist',
                      'analyst': 'Data Analyst', 'data analitics': 'Data Analyst', 'data engineer': 'Data Engineer',
                      'data modeler': 'Data Modeler', 'machine learning engineer': 'ML Engineer',
                      'other scientist': 'Other Scientist', 'director': 'Director', 'na': 'Not Specified' }

df['Job_Title_Sim'] = df['Job_Title_Sim'].astype(str).str.lower().replace(job_title_mapping)
```

```python
# CLEAN Seniority_Level
seniority_mapping = { 'sr': 'Senior', 'jr': 'Junior', 'na': 'Not Specified' }

df['Seniority_Level'] = df['Seniority_Level'].astype(str).str.lower().replace(seniority_mapping)
df['Seniority_Level'] = df['Seniority_Level'].fillna('Not Specified')
```

```python
# CLEAN Employee_Degree
degree_mapping = { 'm': 'Masters', 'p': 'PhD', 'na': 'Not Specified' }

df['Employee_Degree'] = df['Employee_Degree'].astype(str).str.lower().replace(degree_mapping)
```

```python
# CREATE Ownership_Category 
ownership_mapping = { 'Company - Private': 'Corporate', 'Company - Public': 'Corporate',
                      'Subsidiary or Business Segment': 'Corporate', 'Government': 'Government',
                      'Hospital': 'Healthcare', 'Nonprofit Organization': 'Nonprofit',
                      'College / University': 'Education', 'School / School District': 'Education',
                      'Other Organization': 'Other' }

df['Ownership_Category'] = df['Type_Of_Ownership'].map(ownership_mapping).fillna('Not Specified')
```

```python
# REPLACE -1 VALUES 
df['Industry'] = df['Industry'].replace('-1', 'Unknown / Non-Applicable')
df['Sector'] = df['Sector'].replace('-1', 'Unknown / Non-Applicable')
df['Rating'] = df['Rating'].replace(-1, 0)
df['Company_Age'] = df['Company_Age'].replace(-1, 0)
df['Company_Founded_Year'] = df['Company_Founded_Year'].replace(-1, np.nan)
```

```python
# Convert float to integer 
df['Company_Founded_Year'] = df['Company_Founded_Year'].astype('Int64')
```

```python
# Save cleaned data
df.to_csv('data/processed/cleaned_jobs.csv', index=False)
```

### 2. Missing Values Analysis

```python
# Check Missing Values

missing_values = df.isnull().sum()

print(missing_values[missing_values > 0])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Visualization

plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

### 3. Job Role Analysis

```python
# Most Common Job Roles

job_counts = df['Job_Title_Sim'].value_counts()

print(job_counts)

job_counts.plot(kind='bar', figsize=(10,5))
plt.title("Most Common Job Roles")
plt.xlabel("Job Role")
plt.ylabel("Count")
plt.show()
```

### 4. Salary Analysis

```python
# Describe the Avg_Salary
df['Salary_Avg'].describe()
```

```python
#Salary Distribution

plt.figure(figsize=(10,5))

sns.histplot(df['Salary_Avg'], bins=30, kde=True)

plt.title("Salary Distribution")
plt.xlabel("Average Salary")
plt.ylabel("Frequency")

plt.show()
```

### 5. Top Skills Analysis

```python
skills = [
    'Python', 'SQL', 'Excel', 'AWS',
    'Spark', 'TensorFlow', 'Tableau',
    'Power BI'
]

skill_counts = {}

for skill in skills:
    skill_counts[skill] = df[skill].sum()

skill_df = pd.DataFrame({
    'Skill': skill_counts.keys(),
    'Count': skill_counts.values()
})

print(skill_df)

sns.barplot(data=skill_df, x='Skill', y='Count')

plt.title("Most Demanded Skills")
plt.xticks(rotation=45)

plt.show()
```

### 6. Top Hiring Cities

```python
top_cities = df['Job_City'].value_counts().head(10)

print(top_cities)

top_cities.plot(kind='bar')

plt.title("Top Hiring Cities")
plt.xlabel("City")
plt.ylabel("Number of Jobs")

plt.show()
```

### 7. Industry Analysis

```python
# Top Industries Hiring Data Professionals

top_industries = df['Industry'].value_counts().head(10)

print(top_industries)

top_industries.plot(kind='barh', figsize=(10,6))

plt.title("Top Industries")
plt.xlabel("Job Count")

plt.show()
```
